1.   판다스
csv 같은 파일을 쉽게 분석할수 있는 모듈


*   import pandas as pd: 판다스 로딩
*   pd.read_csv("): dataframe으로 로딩
*   type(a): 파일 a의 변수를 반환


*   dataframe['칼럼명'].value_count(): 지정된 칼럼의 데이터값 건수를 반환
*   value_counts(dropna=False):null 값을 포함하여 value_count()적용 시 사용



2.  DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환


*   ndarray->dataframe
pd.DataFrame(a, columns=칼럼명)
*   리스트-> dataframe
pd.DataFrame(리스트 명, columns=칼럼명)


*   딕셔너리-> DataFrame
pd.DataFrame({'칼럼명':[리스트형 또는 ndarray 칼럼 데이터]})
[리스트형 또는 ndarray 칼럼 데이터]: ex. [1, 11]=1행1열에 1, 2행 1열에 11
*   DataFrame->리스트/딕셔너리
-리스트
values로 얻은 ndarray. tolist()
-딕셔너리
a.to_dict('')












3.   DataFrame의 칼럼 데이터 세트 생성과 수정


*   DataFrame[새로운 칼럼명 입력]=c: 뉴 칼럼 생성 및 값 일괄적 할당


4.   DataFrame 데이터 삭제


*   a(데이터프레임 이름).drop('칼럼명', axis=1),: 지정된 칼럼 삭제
*   axis0: 로우방향 축(이땐 칼럼명 지정 안함, 주로 이상치 삭제 경우에 사용)/ axis 1: 칼럼 방향 축



5.   Index 객체


*   a.index: index 객체 추출
*   index객체이름. values: index 객체의 ndarray 값 확인 가능
cf) series 객체에 연산 함수 사용시 index는 제외됨

6.   데이터 셀렉션 및 필터링


*   DataFrame에서 []: 데이터를 가져올 범위(칼럼명) 지정해 가져오기
*   loc[]:  명칭 입력-> 데이터 선택
*   loc[인덱스값 입력(자연수), '칼럼명 입력']
*   iloc[]: 위치(행및 열)  입력-> 데이터 선택
*   iloc[:, :-1]: 처음부터 맨 마지막 칼럼을 제외한 모든 칼럼값 반환
*   iloc[:, -1]: 맨 마지막 칼럼 값 반환











6.   정렬, Agregation 함수, GroupBy 적용


*   sort_values()
- ascending= True 시: 오름차순
- ascending= False 시: 내림차순(기본)
*   .count(): null 값을 반영하지 않은 결과 반환
*   groupby(by= '칼럼명'): 지정한 칼럼명 기준으로 groupby된 것을 반환

7.   결손 데이터 처리하기


*   a.isna(): nan 여부 알려줌
*   a.sum(): isna 후 적용 시 결손 데이터 개수 확인 가능
*   a['교체될 칼럼'].fillna('교체할 데이터'): 결손 데이터 교체





8.   apply lambda 식으로 데이터 가공


*   a(숫자)= lambda x: x**2
숫자의 제곱
cf)숫자 말고 여러개 숫자 사용시 map(lambda x: x**2, 벡터명) 사용

ex

```python
#  나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat =''
  if age<= 5: cat ='Baby'
  elif age<= 12: cat = 'Child'
  elif age<= 18: cat = 'Teenager'
  elif age<= 25: cat = 'Student'
  elif age<= 35: cat = 'Young Adult'
  elif age<= 60: cat = 'Adult'
  else : cat = 'Elderly'

  return cat
# lambda 식에 위에서 생성한 get_category()함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']= titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
