### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
import pandas as pd

url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
df = pd.read_csv(url, sep="\t")
```

```python
csv가 아니라 tab으로 구분된 파일은 sep="\t"인자 추가하기
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = df
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
import seaborn as sns
import matplotlib.pyplot as plt
df1=chipo["item_name"].value_counts().head(5).reset_index()
df1.columns = ["item_name", "count"]
sns.barplot(data=df1, x="item_name", y="count")
plt.title("top items bought")
plt.show()
```

```python
x축인 item_name이 숫자형이 아니라서 범위를 나눌 수 없음-> 숫자 구간 당 도수를 보이는 히스토그램보다 바플롯이 나음
```

```python
top5 = chipo["item_name"].value_counts().head(5).index
chipo_top5 = chipo[chipo["item_name"].isin(top5)]
sns.histplot(x="item_name", data=chipo_top5)
plt.xticks(rotation=45)
plt.show()
```

```python
.index= 라벨(아이템 이름)만 뽑기
reset_index() = 인덱스를 칼럼으로 바꾸기
a.isin(): ()안의 값이 a 안에 있는 만 필터링 (여러 개랑 비교)
==: 하나랑 비교
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
price_map = chipo.groupby("item_name")["item_price"].mean().reset_index()
sns.scatterplot(x="item_price", y="item_name", data=price_map)
plt.show()
```
