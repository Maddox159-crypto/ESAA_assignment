#### 24. Consider an integer vector Z, which of these expressions are legal? (★☆☆)
```python
Z**Z
2 << Z >> 2
Z <- Z
1j*Z
Z/1/1
Z<Z>Z
```

```
Z**Z              <-legal
2 << Z >> 2       <-legal
Z <- Z            <-illegal
1j*Z              <-legal
Z/1/1             <-legal
Z<Z>Z             <-illegal
```

````python
Z**Z              <-legal
2 << Z >> 2       <-legal
<<: 2의 거듭제곱 곱하기   x << n  =  x * 2^n
>>: 2의 거듭제곱 나누기   x >> n = x / 2^n
Z <- Z            <-illegal    (<-는 존재 안함)
1j*Z              <-legal      [1j, 2j, 3j]
Z/1/1             <-legal
Z<Z>Z             <-illegal     (Z<Z) & (Z>Z)
````

#### 36. How to find the closest value (to a given scalar v) in a vector Z? (★★☆)
```python

v = np.random.uniform(0,100)
```
Hint : Coumpute the distances between the each elements of Z and the scalar v. After that, we can see the index of minimum value using `argmin()`.  

````python
array22 = np.arange(100)
array23 = np.random.uniform(0,100)
closest = array22[(np.abs(array22-array23)).argmin()]
print(array23)
print(closest)
````
